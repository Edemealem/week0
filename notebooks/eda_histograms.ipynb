```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

**I am conducting separate analyses for both datasets provided to me.**

**Load data**

1.   benin-malanville




```python
benin_df = pd.read_csv("/content/drive/MyDrive/10Academy/benin-malanville.csv")
benin_df.head(3)
```

**Histograms for Benin**

```python
column = ['GHI', 'DNI', 'DHI', 'WS', 'TModA', 'TModB']

plt.figure(figsize=(10, 6))
for i, col in enumerate(column, 1):
    plt.subplot(2, 3, i)
    plt.hist(benin_df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
