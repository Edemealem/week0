```python
import pandas as pd
import matplotlib.pyplot as pl
from google.colab import drive
drive.mount('/content/drive')
```

**I am conducting separate analyses for both datasets provided to me.**

**Load data**

1.   benin-malanville




```python
benin_df = pd.read_csv("/content/drive/MyDrive/10Academy/benin-malanville.csv")
benin_df.head(3)
```

**Wind Analysis**

```python
wind_speed = benin_df['WS']
wind_direction = benin_df['WD']

#plot radial_bar_plot
radial_bar_plot = pl.figure(figsize=(8, 8))
ax = radial_bar_plot.add_subplot(111, projection='polar')
ax.scatter(wind_direction, wind_speed, s=10, alpha=0.5)
```
