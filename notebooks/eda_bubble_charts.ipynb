Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

**I am conducting separate analyses for both datasets provided to me.**

**Load data**

1.   benin-malanville




```python
benin_df = pd.read_csv("/content/drive/MyDrive/10Academy/benin-malanville.csv")
benin_df.head(3)
```

**Creating a Bubble Chart for Benin**

```python
x = benin_df['GHI']
y = benin_df['Tamb']
z = benin_df['RH']

#bublie plot
plt.figure(figsize=(12, 8))
plt.scatter(x, y, s=z * 10, alpha=0.5, c='blue', edgecolors='w', linewidth=0.5)
plt.title('Bubble Chart: GHI vs. Tamb vs. WS')
plt.xlabel('GHI')
plt.ylabel('Tamb')
plt.colorbar(label='Relative Humidity')
plt.grid()
plt.show();
```
