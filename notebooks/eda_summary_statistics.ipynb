Load data

1.   benin-malanville




```python
benin_df = pd.read_csv("/content/drive/MyDrive/10Academy/benin-malanville.csv")
benin_df.head(3)
```

**Summery Statstics**

```python
benin_df.describe().T
```

Exclude non numeric

```python
benin_df = benin_df.select_dtypes(include=['number'])
```

```python
Mean = benin_df.mean()
Median = benin_df.median()
STD = benin_df.std()
Var = benin_df.var()

#print values
print("\nMean:\n", Mean)
print("\nMedian:\n", Median)
print("\nStandard Deviation:\n", STD)
print("\nVariance:\n", Var)
```
